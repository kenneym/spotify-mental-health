```python
#Written by Soo Park & Matt Kenney

import pandas as pd
import os, sys

#list of all df's of song features
master_list=[]

path = "/Users/Soohwan/Documents/Dartmouth/20-21/20F/COSC 89.20/FinalProject/data/processed-song-data/"
dirs = os.listdir(path)

#import all daily processed csv's and clean them into a single list of dataframes
for file in dirs:
    if (file.endswith('.csv')):
        df = pd.read_csv(path+file)
        df=df.drop(columns=['Track Name','Artist','URL','type','uri','track_href','analysis_url','id'],axis=1)
        master_list.append(df)

#showing the sample
master_list[0].head()
```

```python
import numpy as np

viral_master_avg=pd.DataFrame(columns=["Date","danceability","energy","key","loudness","mode","speechiness","acousticness","instrumentalness","liveness","valence","tempo","duration_ms","time_signature"])

for i in range(len(master_list)):
    danceability=master_list[i]["danceability"].mean()
    energy=master_list[i]["energy"].mean()
    key=master_list[i]["key"].mean()
    loudness=master_list[i]["loudness"].mean()
    mode=master_list[i]["mode"].mean()
    speechiness=master_list[i]["speechiness"].mean()
    acousticness=master_list[i]["acousticness"].mean()
    instrumentalness=master_list[i]["instrumentalness"].mean()
    liveness=master_list[i]["liveness"].mean()
    valence=master_list[i]["valence"].mean()
    tempo=master_list[i]["tempo"].mean()
    duration_ms=master_list[i]["duration_ms"].mean()
    time_signature=master_list[i]["time_signature"].mean()
    df= pd.DataFrame(np.array([[i,danceability,energy,key,loudness,mode,speechiness,acousticness,instrumentalness,liveness,valence,tempo,duration_ms,time_signature]]),columns=["Date","danceability","energy","key","loudness","mode","speechiness","acousticness","instrumentalness","liveness","valence","tempo","duration_ms","time_signature"])
    viral_master_avg=viral_master_avg.append(df,ignore_index=True)

viral_master_avg.head()
```

```python
import matplotlib
import matplotlib.pyplot as plt

plt.plot(viral_master_avg["Date"],viral_master_avg["danceability"])
```
